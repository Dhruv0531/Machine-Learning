```python
import numpy as np
```

```python
def mp_neuron(x, w, threshold):
    # Calculate weighted sum
    weighted_sum = np.dot(x, w)
    # Apply step activation function
    return 1 if weighted_sum >= threshold else 0
```

```python
inputs = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
```

```python
print("AND Gate:")
weights = np.array([1, 1])
threshold = 2

for x in inputs:
    print(f"Input: {x}, Output: {mp_neuron(x, weights, threshold)}")
```

```python
print("\nOR Gate:")
weights = np.array([1, 1])
threshold = 1

for x in inputs:
    print(f"Input: {x}, Output: {mp_neuron(x, weights, threshold)}")
```

```python
print("\nNAND Gate:")
weights = np.array([-2, -2])
threshold = -3

for x in inputs:
    print(f"Input: {x}, Output: {mp_neuron(x, weights, threshold)}")
```

```python
print("\nNOR Gate:")
weights = np.array([-1, -1])
threshold = -1

for x in inputs:
    print(f"Input: {x}, Output: {mp_neuron(x, weights, threshold)}")
```
